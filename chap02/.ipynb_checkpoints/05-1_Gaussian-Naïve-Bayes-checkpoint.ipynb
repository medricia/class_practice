```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
from sklearn.metrics import accuracy_score

np.random.seed(5)
```

# iris 데이터 시각화

```python
# iris 데이터를 불러옴
dataset = load_iris()

#pandas의 데이터프레임으로 데이터를 저장
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df.head()
```

```python
df['target'] = dataset.target   #r에서는 이름으로.. target으로 가져오면 범주화시켜서 가져옴
df.head()
```

```python
# 숫자인 분류값을 이해를 돕기 위해 문자로 변경
df.target = df.target.map({0:'setosa', 1: 'versicolor', 2: 'virginica'})
df
```

```python
df.target.value_counts()
```

## iris 데이터의 분포도 확인
* iris 데이터의 분포도를 확인해 보도록 함 -> iris 데이터가 정규분포를 이루고 있는지 확인(가우시안은 정규분포를 이루어야 하기 때문)  

```python
setosa_df = df[df.target == 'setosa']
versicolor_df= df[df.target == 'versicolor']
virginica_df = df[df.target == 'virginica']
```

```python
# sepal length
ax = setosa_df['sepal length (cm)'].plot(kind='hist')

# kde : 확률밀도함수(y축의 값 : 밀도)
# secondary_y : 보조축에 대한 레이블 출력 여부

#커널밀도 추정(Kernel Density Estimator)

setosa_df['sepal length (cm)'].plot(kind='kde', ax=ax, secondary_y = True, title='setosa sepal length', figsize=(8,4)) 
```

# 데이터를 학습 데이터와 테스트 데이터로 나누기

```python
# 20%를 테스트 데이터로 분류
X_train,X_test,y_train,y_test = train_test_split(dataset.data, dataset.target, test_size=0.2)
```

# Gaussian Naive Bayes 분류하기

```python
model = GaussianNB() #인스턴스 생성
```

```python
# 학습, 정답 데이터 넣어서 학습 시키기
model.fit(X_train, y_train) 

# 테스트 데이터로 모델을 테스트
pred = model.predict(X_test)
```

```python
#정밀도, 재현율, f1 스코어
print(metrics.classification_report(y_test, pred))
```

```python
#정확도
accuracy_score(y_test,pred)
```

## Confusion Matrix

```python
print(metrics.confusion_matrix(y_test, pred))
```
