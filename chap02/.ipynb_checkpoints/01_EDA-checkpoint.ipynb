### - 데이터 시각화
* 데이터 특징을 바탕으로 한 공간에 시각화함으로써 우리는 머신러닝 학습에 필요한 특징과 불필요한 특징을 쉽게 구분지을 수 있다.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#영향력이 있는지 구분하기 어려울때 -> 시각화로 확인

# 스틸, 2점슛 데이터 시각화
sns.lmplot( data=df, x='STL', y='2P', fit_reg = False, hue="Pos", markers=["o", "x"], scatter_kws={"s":150}   ) 

# fit_reg = False : 라인 없이 그리기
# hue = 포지션의 feature값
# markers = 범주값
# scatter_kws = 딕셔너리 타입으로 데이터를 넣어줌. s: 150 -> 산점도 크기

#title
plt.title('STL and 2P in plane')
```

```python
# 두 데이터를 왜 비교했는지... 확인

# 분포 분석
# 범위가 바로 나뉘어 지면 포지션을 바로 나눌 수 있겠지만...
# 대략적으로 왼쪽부근은 c, 오른쪽 부근은 sg... 이지만...
# 애매한 부분에 놓인 경우에는..대부분의 데이터는 경계가 모호함
# 2p는 결과를 결정짓는데 있어서 크게 도움이 되지 않음

# 왜 데이터가 겹치는지? 결정짓지 못하는 데이터인지 확인!
# ->데이터를 잘 분석해오는 것이 중요함
# 스틸 : 공을 가로채는것 -> 공격수와 수비수 입장에서 수비하는 입장에선 모두 스틸함 -> 누구든지 가로채고 누구든지 골을 넣을 수 있음
```

```python
sns.lmplot(data=df, x='AST', y='2P', fit_reg = False, hue="Pos", markers=["o", "x"], scatter_kws={"s":150})
plt.title('AST and 2P in plane')
```

```python
# AST : 득점에 도움을 주는 패스/ 공격수가 더 어시스트를 할 것인가, 수비수가 더 어시스트를 할 것인가...?
```

```python
sns.lmplot(data=df, x='BLK', y='3P', fit_reg = False, hue="Pos", markers=["o", "x"], scatter_kws={"s":150})
plt.title('BLK and 3P in plane')
```

```python
sns.lmplot(data=df, x='TRB', y='3P', fit_reg = False, hue="Pos", markers=["o", "x"], scatter_kws={"s":150})
plt.title('TRB and 3P in plane')

# 현실데이터는 무조건 딱 나누어질 순 없음
# 위의 plot을 보면 비교적 경계가 나뉘어져 있음. 유효한 feature값임
# 특히 분류 알고리즘에 있어서는 더더 중요!
```

# - 데이터 다듬기

```python
# 3P와 BLK, TRB 명확하게 구분됨 -> 머신러닝 유의미
# 나머지 필요없는 피처 제거하기


#분별력이 없는 속성(feature)을 데이터에서 제거
df. drop(['2P','AST','STL'], axis=1, inplace=True)
df.head()
```

# - 데이터 나누기 (학습데이터, 테스트데이터)

```python
# sklearn(프롬프트에서는 scikit-learn 으로 설치)의 train_test_split()을 사용하면 코드 한 줄로 손쉽게 데이터를 나눌 수 있다
```

```python
from sklearn.model_selection import train_test_split
```

```python
#다듬어진 데이터에서 20%를 테스트 데이터로 분류함
train, test = train_test_split(df, test_size=0.2)

print(train.shape)
print(test.shape)

# 70~80% : 예측 모델을 가지고 학습
# 20% : 검증
```

```python
# 학습 데이터의 갯수를 확인
train.shape[0]
```

```python
# 테스트 데이터의 갯수를 확인
test.shape[0]
```

```python
# 학습데이터 저장
```

```python
train.to_csv('data/csv/basketball_train.csv', index=None)
```

```python
# 테스트데이터 저장
```

```python
test.to_csv('data/csv/basketball_test.csv', index=None)
```
