```python
import tensorflow as tf
import numpy as np
```

```python
# open, high, volume, low, close
xy = np.array([[828.659973, 833.450012, 908100, 828.349976, 831.659973],
               [823.02002, 828.070007, 1828100, 821.655029, 828.070007],
               [819.929993, 824.400024, 1438100, 818.97998, 824.159973],
               [816, 820.958984, 1008100, 815.48999, 819.23999],
               [819.359985, 823, 1188100, 818.469971, 818.97998],
               [819, 823, 1198100, 816, 820.450012],
               [811.700012, 815.25, 1098100, 809.780029, 813.669983],
               [809.51001, 816.659973, 1398100, 804.539978, 809.559998]])
```

```python
def min_max_scaler(data):
    numeraor = data - np.min(data, axis=0)
    denominator = np.max(data, axis=0) - np.min(data, axis=0)
    return numeraor/denominator

# 정규화 : 0~1 사이로 만들어줌(0~1사이 범위로 축소), 내 값이 가장 크면 1, 가장 작으면 0으로 만들겠다는 뜻
# 공통된 간격으로 표현

#주성분분석 -> 표준화
#일반데이터(단위가 다른 데이터)-> 정규화
```

```python
data_scaler = min_max_scaler(xy)
data_scaler
```

```python
x_data = data_scaler[:,:-1]
y_data = data_scaler[:,[-1]]
```

```python
tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.Dense(input_dim=x_data.shape[1], units=y_data.shape[1], activation='linear'))
tf.model.compile(loss='mse', optimizer=tf.keras.optimizers.SGD(learning_rate=1e-1))
loss = tf.model.fit(x_data, y_data, epochs=2000)
```

```python
print(loss.history['loss'])
```

```python
predictions = tf.model.predict(x_data)
```

```python
score = tf.model.evaluate(x_data,y_data)
```

```python
print('Prediction: \n', predictions)
print('Cose:\n',score)
```
