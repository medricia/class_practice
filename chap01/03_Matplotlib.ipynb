```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 단순한 그래프 그리기

# 데이터 준비

x = np.arange(0,6,0.1)
print(x)
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
# 그래프
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

```python
plt.plot(x,y1, label="sin")
plt.plot(x,y2,linestyle="--", label="cos") #cos()함수는 점선으로
plt.xlabel("x") #x축이름
plt.xlabel("y") #y축이름
plt.title("sin&cos") #제목출력
plt.legend() #범례출력
plt.show()
```

```python
np.random.seed(1234) #난수를 고정

x = np.arange(10)
y = np.random.rand(10)

plt.plot(x,y)
plt.show()
```

```python
# 3차 함수 f(x) = (x-2)x(x+2)

def f(x) :
    return (x-2)*x*(x+2)

print(f(0))
print(f(-2))
print(f(2))
```

```python
# x값에 대해 ndarray 배열이며 각각에 대응한 f를 한꺼번에 ndarray로 돌려주기

print(f(np.array([1,2,3])))
print(type(f(np.array([1,2,3])))) #<class 'numpy.ndarray'>
```

```python
# 그래프를 그리는 xdml qjadnlfmf -3~3까지로 하고, 간격은 0.5
# (간격이 좁을 수록 부드러운 모양의 그래프가 그려짐)

x = np.arange(-3,3,0.5)
plt.plot(x, f(x))
plt.show()
```

```python
# 그래프를 장식
def f2(x,w) :
    return (x-w) * x * (x+2) #함수 정의
```

```python
#x를 정의

x = np.linspace(-3, 3 , 100) #-3~3 범위에서 균등하게 100분할해줌(100개의 값 return)
```

```python
#차트 묘사
plt.plot(x,f2(x,2), color='black', label='w=2')
plt.plot(x,f2(x,1), color='blue', label='w=1')
plt.legend(loc="upper left")
plt.ylim(-15,15) #y축 범위
plt.title('(x-w)x(x+2)')
plt.xlabel('x')
plt.ylabel('f2(x,w)')
plt.grid(True) #눈금 표시
plt.show()
```

```python
# 그래프를 여러 개 보여주기
plt.figure(figsize=(10,3)) #전체 영역의 크기를 지정
plt.subplots_adjust(wspace=0.5, hspace=0.5) #그래프의 간격을 지정

for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.title(i+1)
    plt.plot(x, f2(x,i))
    plt.ylim(-20,20)
    plt.grid(True)
    
plt.show()
```
