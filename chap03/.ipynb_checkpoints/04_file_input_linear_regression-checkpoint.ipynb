```python
import tensorflow as tf
import numpy as np
```

```python
# 데이터 읽어오기
data = np.loadtxt('../data/data-01-test-score.csv', delimiter=',', dtype=np.float32) #../ 한단계 위로 올라가는 delimiter:구분단위
x_data = data[ : , 0:-1]
y_data = data[ : , [-1]]
# x축은 마지막의 전까지, y축은 마지막 축만
print(x_data, '\nx_data.shape', x_data.shape)
print(y_data, '\ny_data.shape', y_data.shape)

# numpy의 array로 읽어오고 있음
```

```python
tf.model = tf.keras.Sequential()
#입력데이터의 feature

#선호 코드
tf.model.add(tf.keras.layers.Dense(input_dim=3 ,units=1, activation='linear'))
#tf.model.add(tf.keras.layers.Dense(input_dim=3 ,units=1))
#tf.model.add(tf.keras.layers.Activation('linear'))
# 생략 (선형이 가장 default 값이기 때문에)
```

```python
tf.model.summary()
```

```python
tf.model.compile(loss='mse', optimizer=tf.keras.optimizers.SGD(learning_rate=1e-5))
```

```python
tf.model.fit(x_data, y_data, epochs=2000)
```

```python
tf.model.layers[0].get_weights()
```

```python
y_predict = tf.model.predict([[60., 70., 80.], [73, 80, 75]])
y_predict
```
