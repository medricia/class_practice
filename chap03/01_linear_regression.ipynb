# 신경망 구현 순서
1. Sequential 모형 클래스 객체 생성
2. add 메소드로 레이어 추가
 - 입력단부터 순차적으로 추가한다
 - 레이어는 출력 뉴런 갯수를 첫번째 인수로 받는다
 - 최초의 레이어는 input_dim 인수로 입력크기를 설정해야 한다
 - activation 인수로 활성화 함수 설정
3. compile 메소드로 모형 완성
 - loss 인수로 비용함수 설정
 - optimizer 인수로 최적화 알고리즘 설정
 - metrics 인수로 트레이닝 단계에서 기록할 성능 기준 설정
4. fit 메소드로 트레이닝
 - epochs로 에포크(epoch) 횟수 설정

```python
# tensorflow 설치
# conda install tensorflow

# 텐서플로우 설치 확인
import tensorflow as tf

print(tf.__version__)
```

```python
# H(x) = wx+ b : w=-1, b=1

# 피처가 아닌 4개의 데이터가 있는것임, 각각 1개씩
x_train = [1,2,3,4]
y_train = [0,-1,-2,-3]
```

```python
tf.model = tf.keras.Sequential()
```

```python
#input_dim : input shape, units=out shape
tf.model.add(tf.keras.layers.Dense(input_dim=1, units=1))   # Dense 세포삽입(추가)
```

```python
# SGD(Stochastic Gradient Descendent), lr(learning_rate)
# Stochastic -> 훈련데이터가 많을때, step을 전부 활용x, 랜덤하게 선택된 것으로 
# 손실함수의 값* lr의 값을 뺀 값
# lr 클 때 : 값 핑퐁하다가 발산할 수 있음/ lr작을때 : 학습하다 최적값 찾지 못하고 학습 끝나버림
# 0.1~0.01에서 시작함
#lr : 정확도 -> 초매개변수

sgd = tf.keras.optimizers.SGD(learning_rate=0.1)
```

```python
#모형 완성 : complie

#mse : mean square error (평균제곱오차)-> 손실함수에 mse
#오차(잔차)  

tf.model.compile(loss='mse', optimizer=sgd)
```

```python
tf.model.summary()
```

```python
tf.model.fit(x_train, y_train, epochs=200) #200번 반복하면서 학습하라는 뜻, 손실함수가 최소가 되도록
```

```python
# 학습을 통해 구한 w와 b의 값
tf.model.layers[0].get_weights()

# 실수값의경우 딱 떨어지게 나오지 않음(스텝 이동 할때마다 오차가 발생함)
# w 값은 [[값]] , b 값은 [값] 으로 반환됨
```

```python
#예측해보기
import numpy as np
y_predict = tf.model.predict(np.array([5]))
y_predict
```
